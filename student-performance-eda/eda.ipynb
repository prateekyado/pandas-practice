```python
df["Science"] = pd.to_numeric(df["Science"], errors="coerce")
```

```python
df.fillna({
    "Science" : 0,
    "English" : 0,
    "Attendance" : df["Attendance"].mean()
}, inplace = True)
```

```python
df
```

```python
df["JoinDate"] = pd.to_datetime(df["JoinDate"])
```

```python
df
```

```python
df["Total"] = df[["Math", "Science", "English"]].sum(axis = 1)
```

```python
df["Average"] = df["Total"]/3
```

```python
df["Performance"] = df["Average"].apply(
    lambda x: "Excellent" if x >= 90 else "Good" if x >= 50 else "Failed" 
)
```

```python
df[df["Attendance"] < 90]
```

```python
df[df["Performance"] == "Excellent"]
```

```python
df.groupby("Gender")["Average"].mean()
```

```python
df.groupby("Gender")[["Math","Science","English"]].mean()
```

```python
df["Performance"].value_counts()
```

```python
df.loc[df["Average"] > 80, ["Name","Average","Performance"]]
```

```python
df.rename(columns={"Attendance":"Attendance_Percentage"}, inplace=True)
```

```python
df.to_csv("cleaned_students.csv", index=False)
```
